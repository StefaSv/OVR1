```python
import math

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

datafrarme = pd.read_csv("GOOG.csv",sep=',')

datafrarme = datafrarme["Open"]


plt.plot(list(range(0, 251)), datafrarme, "o", color="navy")
plt.plot(list(range(0, 251)), datafrarme, color="navy", alpha=0.5, lw=2)

plt.grid(True)
plt.show()




```

```python
plt.hist(datafrarme, bins=25)

plt.show()
```

```python
plt.boxplot(datafrarme)

plt.show()
```

```python
print("Среднее значение:      ", datafrarme.mean())
print("Минимальное значение:  ", datafrarme.min())
print("Максисвальное значение:", datafrarme.max())
print("Дисперсия:             ", datafrarme.var())
```
