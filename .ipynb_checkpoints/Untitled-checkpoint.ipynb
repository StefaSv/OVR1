```python
#Импорт библиотек
import math
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

#Чтение файла
datafrarme = pd.read_csv("GOOG.csv",sep=',')

#Будем орассматривать столбец Open
datafrarme = datafrarme["Open"]

#создание фигуры и плота
fig, ax = plt.subplots(figsize=(12, 8)) 

#Строим график по точким и линией
ax.plot(list(range(0, 251)), datafrarme, "o", color="navy", markersize=2)
ax.plot(list(range(0, 251)), datafrarme, color="navy", alpha=0.5, lw=2)

#добавление сетки
plt.grid(True)

#отрисовка плота
plt.show()




```

```python
fig = plt.figure(4)

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

#создание гистограммы с разным количеством столбцов
ax1.hist(datafrarme, round(math.log2(251))+1, label = "Формула Стерджейса")
ax2.hist(datafrarme, bins=20, label = "bins=10")
ax3.hist(datafrarme, bins=30, label = "bins=20")
ax4.hist(datafrarme, bins=40, label = "bins=30")

plt.show()
```

```python
#построение боксплота
plt.boxplot(datafrarme)

plt.show()
```

```python
#рассчёт статистических характеристик
print("Среднее значение:      ", datafrarme.mean())
print("Минимальное значение:  ", datafrarme.min())
print("Максисвальное значение:", datafrarme.max())
print("Дисперсия:             ", datafrarme.var())
```
